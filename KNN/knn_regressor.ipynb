```python
import pandas as pd
import numpy as np
```

```python
class KNNRegressor:

    def __init__(self,k):
        if k<1:
            raise ValueError("K value must be greater than one/odd numbers preferred")
        self.k=k

    def euclidian_distance(self,X,Y):
        return np.sqrt(np.sum((X - Y) ** 2))

    def fit(self,X_train:pd.DataFrame,Y_train:pd.DataFrame):
        # KNN just stores the dataset during training (lazy algo)
        self.X_train=X_train
        self.Y_train=Y_train

        return self

    def predict(self,X_test:pd.DataFrame)->pd.DataFrame:

        X_predict=X_test.copy()
        predictions=[]
        for test_idx in range(X_predict.shape[0]):
            X_test_values=X_predict.iloc[test_idx].values
            result=[]
            for row in range(self.X_train.shape[0]):
                X_train_values=self.X_train.iloc[row].values
                distances=self.euclidian_distance(X_train_values,X_test_values)
                result.append((row,distances))

            sorted_distances=sorted(result,key=lambda x:x[1])
            k_distances=sorted_distances[:self.k]

            y_val=[]
            for index,values in k_distances:
                y_val.append(self.Y_train.iloc[index])

            prediction=round(sum(y_val)/len(y_val),2)
            predictions.append(prediction)

        return predictions

```

```python
train_df=pd.DataFrame({
    "GPA":[7.0,6.7,6.5,5.95,5.75,5.5],
    "IQ":[100,120,90,78,89,79]
})

package=[9,10,12,8,5,6]
y_train=pd.Series(package)

test_df=pd.DataFrame({
    "GPA":[7.0,5.4,7.0,4.3],
    "IQ":[120,40,150,94]
})
```

```python
knn=KNNRegressor(2)
```

```python
knn.fit(train_df,y_train)
```

```python
knn.predict(test_df)
```
