```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
class MyLinearRegression:

    def __init__(self)->None:
        self.slope=None
        self.intercept=None

    def fit(self,X_train,Y_train):

        numerator=0
        denominator=0

        for rows in range(X_train.shape[0]):
            numerator=numerator+(X_train.iloc[rows]-X_train.mean())*(Y_train.iloc[rows]-Y_train.mean())
            denominator=denominator+(X_train.iloc[rows]-X_train.mean())*(X_train.iloc[rows]-X_train.mean())

        self.slope=round(numerator/denominator,3)
        self.intercept=round(Y_train.mean()-self.slope*X_train.mean(),3)

        print(f"Slope:{self.slope},Intercept:{self.intercept}")
    
    def predict(self,X_test):
        result=[]

        for rows in X_test:
            y=self.slope*rows+self.intercept
            result.append(y)

        return result
    
    def evaluate(self,prediction,Y_test):
        error=0
        
        for i in range(len(Y_test)):
            error+=(prediction[i]-Y_test.iloc[i])**2

        return round(error/len(Y_test),3)
```

```python
students=pd.read_csv("placement.csv")
```

```python
X=students['cgpa']
Y=students['package']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=10,test_size=0.2)
```

```python
X_train
```

```python
lr=MyLinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
prediction=lr.predict(X_test)
```

```python
lr.evaluate(prediction,Y_test)
```
