```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
```

```python
class MyStandardization:

    def __init__(self):
        self.mean=None
        self.sd=None

    def fit(self,X_train):
        self.mean=X_train.mean()
        self.sd=X_train.std()

        print(f"The mean is:{self.mean}")
        print(f"The standard deviation is:{self.sd}")

    def transform(self,X_df):
        X_df=(X_df-self.mean)/self.sd
        return X_df
    
    def fit_transform(self,df):
        self.fit(df)
        return self.transform()
```

```python
class Normalization:

    def __init__(self):
        self.min=None
        self.max=None

    def fit(self,X_train):
        self.max=X_train.max()
        self.min=X_train.min()

        print(f"The max value is:{self.max}")
        print(f"The min value is:{self.min}")

    def transform(self,X_df):
        X_df=(X_df-self.min)/(self.max-self.min)
        return X_df
    
    def fit_transform(self,df):
        self.fit(df)
        return self.transform()
```
